```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Parmetros iniciais
x = 51
y = 49
q = 2
n_lancamentos = 20000

# Contadores de bolas recebidas
bolas_recebidas_X = 0
bolas_recebidas_Y = 0

# Histrico do nmero de bolas em cada cesto
historico_X = [x]
historico_Y = [y]

# Simulao Monte Carlo
for _ in range(n_lancamentos):
    prob_X = (x ** q) / (x ** q + y ** q)
    prob_Y = (y ** q) / (x ** q + y ** q)

    if np.random.rand() < prob_X:
        x += 1
        bolas_recebidas_X += 1
    else:
        y += 1
        bolas_recebidas_Y += 1

    historico_X.append(x)
    historico_Y.append(y)

# Determinar o valor de c
c = min(bolas_recebidas_X, bolas_recebidas_Y)

# Determinar qual cesto recebeu mais bolas
if bolas_recebidas_X > bolas_recebidas_Y:
    cesto_mais_bolas = "X"
else:
    cesto_mais_bolas = "Y"

# Plot do grfico
plt.figure(figsize=(12, 6))
plt.plot(historico_X, label="Cesto X")
plt.plot(historico_Y, label="Cesto Y")
plt.xlabel("ndice da jogada")
plt.ylabel("Nmero de bolas")
plt.title("Evoluo do nmero de bolas em cada cesto")
plt.legend()
plt.grid()
plt.show()

```

```python
# Resultados finais
print(f"Valor de c: {c}")
print(f"Cesto que recebeu mais bolas: {cesto_mais_bolas}")
```

```python
def simular_lancamentos(n_lancamentos, x_inicial, y_inicial, q):
    x = x_inicial
    y = y_inicial

    bolas_recebidas_X = 0
    bolas_recebidas_Y = 0

    for _ in range(n_lancamentos):
        prob_X = (x ** q) / (x ** q + y ** q)
        prob_Y = (y ** q) / (x ** q + y ** q)

        if np.random.rand() < prob_X:
            x += 1
            bolas_recebidas_X += 1
        else:
            y += 1
            bolas_recebidas_Y += 1

    return x, y, bolas_recebidas_X, bolas_recebidas_Y

# Parmetros iniciais
x_inicial = 51
y_inicial = 49
q = 2
n_lancamentos = 20000
n_simulacoes = 100

# Contador para verificar quantas vezes X termina com mais bolas
x_vence = 0

for _ in range(n_simulacoes):
    x_final, y_final, _, _ = simular_lancamentos(n_lancamentos, x_inicial, y_inicial, q)
    if x_final > y_final:
        x_vence += 1

frequencia_x_vence = x_vence / n_simulacoes

# Exibir resultado
print(f"Frequncia com que o cesto X termina com mais bolas: {frequencia_x_vence:.2f}")
```
